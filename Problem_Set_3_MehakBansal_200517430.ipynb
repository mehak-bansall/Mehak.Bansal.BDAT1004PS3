```python
# Step 1
import numpy as np
import pandas as pd
```

```python
# Step 2 and Step 3
users = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user", sep = '|')
users
```

```python
# Step 4
mean_age_per_occupation = users.groupby('occupation').age.mean()
mean_age_per_occupation
```

```python
# Step 5
def gender_count (i):
    if i == 'M':
        return 1
    if i == 'F':
        return 0
    
users['Gender_For_Male_Ratio'] = users['gender'].apply(gender_count)


Male_ratio_per_occupation = users.groupby('occupation').Gender_For_Male_Ratio.sum() / users.occupation.value_counts() * 100 
# Sorting
Male_ratio_per_occupation.sort_values(ascending = False)
```

```python
# Step 6
users.groupby('occupation').age.agg(['min', 'max'])
```

```python
# Step 7
users.groupby(['occupation','gender']).mean('age')
```

```python
#Step 8
Gender_ocupation = users.groupby(['occupation', 'gender']).agg({'gender': 'count'})
Count_Occupation = users.groupby(['occupation']).count()
occupation_gender = Gender_ocupation.div(Count_Occupation, level = "occupation")
occupation_gender.loc[:, 'gender']
```

```python
# Step 1
import pandas as pd
import numpy as np
import datetime
```

```python
# Step 2 and Step 3
data = pd.read_csv(r'C:\Users\Pavilion\Downloads\wind.csv', sep = '\s+', parse_dates = [[0,1,2]])
data
```

```python
# Step 4
def Fixing_Year(i):
  year = i.year - 100 if i.year > 1989 else i.year
  return datetime.date(year, i.month, i.day)

data['Yr_Mo_Dy'] = data['Yr_Mo_Dy'].apply(Fixing_Year)
data.head()
```

```python
# Step 5
data['Yr_Mo_Dy'] = pd.to_datetime(data['Yr_Mo_Dy'])
data = data.set_index('Yr_Mo_Dy')
data.head()
# data['Yr_Mo_Dy'].dtype
```

```python
# Step 6
Missing_Values = data.isnull().sum()
Missing_Values
```

```python
# Step 7
Total_Non_Missing_Values = data.notnull().sum()
Total_Non_Missing_Values 
```

```python
# Step 8
Mean_windspeeds_of_the_windspeeds = data.sum().sum() / data.notna().sum().sum()
Mean_windspeeds_of_the_windspeeds
```

```python
# Step 10

min = data.min(axis=1)
max = data.max(axis=1)
mean = data.mean(axis=1)
Standard_deviation = data.std(axis=1)
day_stats = pd.DataFrame({'Min':min, 'Max':max, 'Mean':mean, 'Standard Deviation':Standard_deviation})
day_stats
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Marriages_Divorces = pd.read_csv(r'C:\Users\Pavilion\Downloads\us-marriages-divorces-1867-2014.csv')
Marriages_Divorces
```

```python
columns = Marriages_Divorces.columns.drop(['Year','Marriages','Divorces','Population'])

x_data = Marriages_Divorces['Year'].values

fig, ax = plt.subplots()
for column in columns:
    ax.plot(x_data, Marriages_Divorces[column])
ax.set_title('Number of marriages and divorces per capita in the U.S. \n between 1867 and 2014')
ax.legend(columns)
ax.set_xlabel('Year')
ax.set_ylabel('Number Of Marriages Per Capita \n and Divorces Per Capita')
```

# Question 7

```python
Marriages_Divorces = Marriages_Divorces[Marriages_Divorces['Year'].apply(lambda x: x in [1900, 1950, 2000])]
Marriages_Divorces
```

```python
x_data = Marriages_Divorces['Year'].values
x = np.arange(len(Marriages_Divorces['Year'].values))

plt.bar(x_data, Marriages_Divorces['Marriages_per_1000'].values, width=2.0, label = 'Marriages_per_1000')
plt.bar(x_data, Marriages_Divorces['Divorces_per_1000'].values, width=4.0, label = 'Divorces_per_1000')
plt.legend()
plt.title('Number of marriages and divorces per capita in the U.S. between 1900, 1950, and 2000 \n ')
plt.xlabel('Year')
plt.ylabel('Number Of Marriages Per Capita \n and Divorces Per Capita')
plt.show()
```
